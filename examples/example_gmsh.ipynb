## DEVELOPING AN AIRFOIL MESH FROM GMSH 4.13.1

```python
import sys
sys.path.append('/home/m.jaraiz/repos/pyLowOrder')
from FotR import ARAGORN
from FotR.solver_tools import GmshCase
```

```python
runner = ARAGORN(timeout=None)
script_path = '/home/m.jaraiz/Documentos/DATASETS/data_TIFON/gmsh_files/script_geo_v4.geo'
```

```python
result = runner.run(
    script=script_path,
    program="gmsh",
    stdin_mode=False,
    extra_args=["-3", "-o", "mesh.msh"],
    n_cpus=8
)
```

```python

case = (
    GmshCase(script_path, workdir="./case001")
    .set_dimension(3)
    .set_format("msh2")
)

result = case.run(runner, run_id="airfoil_001")

print(result)
```
