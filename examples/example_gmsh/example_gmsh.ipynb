## DEVELOPING AN AIRFOIL MESH FROM GMSH 4.13.1

```python
from FotR import ARAGORN
from FotR.solver_tools import GmshCase
import shutil

try:
    shutil.rmtree('./meshes')
except:
    print('No está')
    
```

```python
runner = ARAGORN(
    timeout=240,
    max_ram_gb=10,
)
script_path = './gmsh_files/script_geo_v4.geo'
# script_path = '/home/m.jaraiz/Documentos/DATASETS/data_TIFON/gmsh_files/script_geo_v4.geo'

case = (
    GmshCase(
        script_path,
        workdir="./meshes",
        mesh_name="mesh_f22_transonic.msh"
    )
    .set_dimension(3)
    .set_format("msh2")
    .set_n_cpus(100)
)
# print(case.get_executable())
```

```python
result = case.run(
    runner, run_id="airfoil_001")

print(result)
```

```python
case.plot(mode="auto")
```
