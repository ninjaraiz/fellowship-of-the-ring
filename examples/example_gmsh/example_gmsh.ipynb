```python
runner = ARAGORN(timeout=None)
script_path = './gmsh_files/script_geo_v5.geo'
# script_path = '/home/m.jaraiz/Documentos/DATASETS/data_TIFON/gmsh_files/script_geo_v4.geo'

case = (
    GmshCase(script_path, workdir="./case001", mesh_name="mesh.msh")
    .set_dimension(3)
    .set_format("msh2")
)
# print(case.get_executable())
```

```python
result = case.run(runner, run_id="airfoil_001")

print(result)
```
