```python
import numpy as np
from typing import Union
from FotR import SAM
def group_to_new_mesh_3(
    db: 'FRODO',
    id_group_src: str,
    new_group_id: str,
    new_mesh: dict,
    vars: Union[str, list[str]] = 'all',
    method: str = 'idw',
    k: int = 4,
    ):

    data_src = db.data_dict[f'CADGroup_{id_group_src}']
    coord_src = data_src["Coord"]
    conec_src = data_src.get("Conec", None)
    vars_src = data_src["Vars"]

    if "Coord" not in new_mesh:
        raise ValueError("new_mesh debe contener 'Coord'")

    coord_dst = new_mesh["Coord"]
    conec_dst = new_mesh.get("Conec", None)

    new_group_key = f'CADGroup_{new_group_id}'
    db.data_dict[new_group_key] = {}

    # copiar malla
    for key, value in new_mesh.items():
        db.data_dict[new_group_key][key] = value

    # copiar condiciones de vuelo
    if "FlCc" in data_src:
        db.data_dict[new_group_key]["FlCc"] = data_src["FlCc"]

    db.data_dict[new_group_key]["Vars"] = {}

    # -------------------------
    # KDTree cache
    # -------------------------
    if method == "idw":
        from scipy.spatial import cKDTree
        tree = cKDTree(coord_src)
    else:
        tree = None

    # -------------------------
    # Loop stages
    # -------------------------
    for stage, stage_data in vars_src.items():

        db.data_dict[new_group_key]["Vars"][stage] = {}

        if vars == 'all':
            selected_vars = [
                v for v in stage_data.keys()
                if v not in ("GlobalNumber", "CADGroupID")
            ]
        else:
            selected_vars = vars

        var_list = []
        shapes = []
        valid_names = []

        for var_name in selected_vars:
            if var_name not in stage_data:
                continue

            v = stage_data[var_name]

            if v.ndim == 1:
                v = v[:, None]

            var_list.append(v)
            shapes.append(v.shape[1])
            valid_names.append(var_name)

        if not var_list:
            continue

        var_src_stack = np.hstack(var_list)

        # -------------------------
        # Interpolación
        # -------------------------
        if method == "idw":
            var_dst_stack = SAM.Weapons._interpolate_idw_tree(
                tree, coord_src, var_src_stack, coord_dst, k=k
            )

        elif method == "griddata":
            var_dst_stack = SAM.Weapons._interpolate_griddata(
                coord_src, var_src_stack, coord_dst
            )

        elif method == "pyvista":
            if conec_src is None or conec_dst is None:
                raise ValueError("PyVista requiere 'Conec' en ambas mallas")

            var_dst_stack = SAM.Weapons._interpolate_pyvista(
                coord_src, conec_src,
                var_src_stack,
                coord_dst, conec_dst
            )

        else:
            raise ValueError(f"Método desconocido: {method}")

        # -------------------------
        # Unstack
        # -------------------------
        idx = 0
        for var_name, dim in zip(valid_names, shapes):
            db.data_dict[new_group_key]["Vars"][stage][var_name] = \
                var_dst_stack[:, idx:idx+dim]
            idx += dim

    return db

new_mesh = {
    "Coord": db.data_dict['CADGroup_3']['Coord'],
    "Conec": db.data_dict['CADGroup_3']['Conec'],
    "NodeCoord": db.data_dict['CADGroup_3']['NodeCoord'],
    # lo que tengas
}

db = group_to_new_mesh_3(
    db=db_prueba,
    id_group_src='3',
    new_group_id='3_interp',
    new_mesh=new_mesh,
    method='idw',
    k=8
)
db_prueba.summary_data()
```

```python
from typing import Union, List
import numpy as np
from FotR import SAM

def group_to_new_mesh_2(
    db: 'FRODO',
    id_group_src: str,
    new_group_id: str,
    new_mesh: dict,
    vars: Union[str, List[str]] = 'all',
    method: str = 'idw',
    k: int = 4,
):
    """
    Proyecta Vars a nueva malla + vectorizado + copia condiciones de vuelo.
    """

    # -------------------------
    # Datos origen
    # -------------------------
    data_src = db.data_dict[f'CADGroup_{id_group_src}']
    coord_src = data_src["Coord"]
    vars_src = data_src["Vars"]

    # -------------------------
    # Datos destino
    # -------------------------
    if "Coord" not in new_mesh:
        raise ValueError("new_mesh debe contener 'Coord'")

    coord_dst = new_mesh["Coord"]

    # -------------------------
    # Crear nuevo grupo
    # -------------------------
    new_group_key = f'CADGroup_{new_group_id}'
    db.data_dict[new_group_key] = {}

    # copiar geometría
    for key, value in new_mesh.items():
        db.data_dict[new_group_key][key] = value

    # ✅ copiar condiciones de vuelo
    if "FlCc" in data_src:
        db.data_dict[new_group_key]["FlCc"] = data_src["FlCc"]

    # inicializar Vars
    db.data_dict[new_group_key]["Vars"] = {}

    # -------------------------
    # Loop por stages
    # -------------------------
    for stage, stage_data in vars_src.items():

        db.data_dict[new_group_key]["Vars"][stage] = {}

        # selección de variables
        if vars == 'all':
            selected_vars = [
                v for v in stage_data.keys()
                if v not in ("GlobalNumber", "CADGroupID")
            ]
        else:
            selected_vars = vars

        # -------------------------
        # STACK de variables
        # -------------------------
        var_list = []
        shapes = []

        for var_name in selected_vars:
            if var_name not in stage_data:
                continue

            v = stage_data[var_name]

            # asegurar 2D
            if v.ndim == 1:
                v = v[:, None]

            var_list.append(v)
            shapes.append(v.shape[1])

        if not var_list:
            continue

        var_src_stack = np.hstack(var_list)  # (N, total_features)

        # -------------------------
        # Interpolación única
        # -------------------------
        if method == "idw":
            var_dst_stack = SAM.Weapons._interpolate_idw(
                coord_src, var_src_stack, coord_dst, k=k
            )

        elif method == "griddata":
            var_dst_stack = SAM.Weapons._interpolate_griddata(
                coord_src, var_src_stack, coord_dst
            )

        else:
            raise ValueError(f"Método desconocido: {method}")

        # -------------------------
        # UNSTACK
        # -------------------------
        idx = 0
        for var_name, dim in zip(selected_vars, shapes):
            db.data_dict[new_group_key]["Vars"][stage][var_name] = \
                var_dst_stack[:, idx:idx+dim]
            idx += dim

    return db

new_mesh = {
    "Coord": db.data_dict['CADGroup_3']['Coord'],
    "Conec": db.data_dict['CADGroup_3']['Conec'],
    "NodeCoord": db.data_dict['CADGroup_3']['NodeCoord'],
    # lo que tengas
}

db = group_to_new_mesh_2(
    db=db_prueba,
    id_group_src='3',
    new_group_id='3_interp',
    new_mesh=new_mesh,
    method='idw',
    k=8
)
```

```python
from typing import Union, List
from FotR import SAM
def group_to_new_mesh_1(
    db: 'FRODO',
    id_group_src: str,
    new_group_id: str,
    new_mesh: dict,
    vars: Union[str, List[str]] = 'all',
    method: str = 'idw',
    k: int = 4,
    ):
    """
    Proyecta las Vars de un grupo existente a una nueva malla y crea un nuevo CADGroup.
    
    new_mesh debe contener al menos:
        - "Coord"
    Opcional:
        - "Conec", "NodeCoord", etc.
    """

    # -------------------------
    # Datos origen
    # -------------------------
    data_src = db.data_dict[f'CADGroup_{id_group_src}']
    coord_src = data_src["Coord"]

    vars_src = data_src["Vars"]

    # -------------------------
    # Datos destino (nueva malla)
    # -------------------------
    if "Coord" not in new_mesh:
        raise ValueError("new_mesh debe contener 'Coord'")

    coord_dst = new_mesh["Coord"]

    # -------------------------
    # Crear nuevo grupo
    # -------------------------
    new_group_key = f'CADGroup_{new_group_id}'

    db.data_dict[new_group_key] = {}

    # copiar geometría completa
    for key, value in new_mesh.items():
        db.data_dict[new_group_key][key] = value

    # inicializar Vars
    db.data_dict[new_group_key]["Vars"] = {}

    # -------------------------
    # Loop por stages
    # -------------------------
    for stage, stage_data in vars_src.items():

        db.data_dict[new_group_key]["Vars"][stage] = {}

        # selección de variables
        if vars == 'all':
            selected_vars = list(stage_data.keys())
        else:
            selected_vars = vars

        for var_name in selected_vars:

            if var_name not in stage_data:
                continue

            # ⚠️ evitar variables no físicas
            if var_name in ("GlobalNumber", "CADGroupID"):
                continue

            var_src = stage_data[var_name]

            # -------------------------
            # Interpolación
            # -------------------------
            if method == "idw":
                var_dst = SAM.Weapons._interpolate_idw(
                    coord_src, var_src, coord_dst, k=k
                )

            elif method == "griddata":
                var_dst = SAM.Weapons._interpolate_griddata(
                    coord_src, var_src, coord_dst
                )

            else:
                raise ValueError(f"Método desconocido: {method}")

            # -------------------------
            # Guardar
            # -------------------------
            db.data_dict[new_group_key]["Vars"][stage][var_name] = var_dst

    return db
    
new_mesh = {
    "Coord": db.data_dict['CADGroup_3']['Coord'],
    "Conec": db.data_dict['CADGroup_3']['Conec'],
    "NodeCoord": db.data_dict['CADGroup_3']['NodeCoord'],
    # lo que tengas
}

db = group_to_new_mesh_1(
    db=db_prueba,
    id_group_src='3',
    new_group_id='3_interp',
    new_mesh=new_mesh,
    method='idw',
    k=8
)
```
